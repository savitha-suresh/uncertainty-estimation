### Load into Dataloader

```python
class FashionMNISTDataset(Dataset):
    def __init__(self, annotations_file, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file)
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        label = self.img_labels.iloc[idx, 0]
        image = self.img_labels.iloc[idx, 1:].values
        image = torch.tensor(image, dtype=torch.float32).view(1, 28, 28)
        image = image / 255.0

        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        
        return image, label
```

```python
train_csv_path = os.path.join(csv_dir, "train_data.csv")
test_csv_path = os.path.join(csv_dir, "test_data.csv")
```

```python
train_dataset = FashionMNISTDataset(annotations_file=train_csv_path)
test_dataset = FashionMNISTDataset(annotations_file=test_csv_path)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=64, shuffle=False)
print(f" DataLoader process complete! Training samples: {len(train_dataset)} Test samples: {len(test_dataset)}")
```
